```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import cross_val_score

train = pd.read_csv('train[1].csv')
test = pd.read_csv('test[1].csv')

features = ['Pclass', 'Sex', 'Age', 'Fare']

X = train[features].copy()
X_test = test[features].copy()

X['Age'].fillna(X['Age'].median(), inplace=True)
X_test['Age'].fillna(X['Age'].median(), inplace=True)
X_test['Fare'].fillna(X_test['Fare'].median(), inplace=True)

X = pd.get_dummies(X)
X_test = pd.get_dummies(X_test)
X_test = X_test.reindex(columns=X.columns, fill_value=0)

model = xgb.XGBClassifier(max_depth=3, learning_rate=0.1, n_estimators=100, random_state=42)

cv_scores=cross_val_score(model, X, train['Survived'], cv=5)
print('Accuracy:',cv_scores)

model.fit(X, train['Survived'])

pred = model.predict(X_test)
pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': pred}).to_csv('submission_4feat.csv', index=False)
print("submission_4feat.csv ready! Upload to Kaggle")
```
